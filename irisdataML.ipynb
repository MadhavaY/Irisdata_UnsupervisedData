```python
import pandas as pd
df = pd.read_csv("D:/iris.csv")
df
```

```python
x = df['petal.length']
y = df['petal.width']

data = list(zip(x.values,y.values))
data
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
model = KMeans(n_clusters=4)
model.fit(data)
```

```python
plt.scatter(x,y, c= model.labels_)
plt.show
```
